```python
def select_cars(katalog, param, criteria):
    selected_cars = []
    for car in katalog:
        specification = list(car.values())[0]
        if specification[param] == criteria:
            selected_cars.append(list(car.keys()))
    if selected_cars:
        return selected_cars
    else:
        return "Таких машин нет"
```
